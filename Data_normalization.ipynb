```python
def is_position(df, ind, position):
    e=df['positions'].iloc[ind]
    res=False
    for pos in position:
        is_pos= any(item['position']['code'] == pos for item in e)
        res=res or is_pos

    return res
```

```python
#reading and filtering the file according the the positions
def readfile(path, poslist):
    with open(path, 'r') as f:
        rd = f.read()
    obj = json.loads(rd)
    dfw = pd.json_normalize(obj)
    dfw = pd.json_normalize(dfw['players'])
    l=dfw.shape[1]
    df=pd.read_csv('empty_df.csv')
    for i in range(l):
        dftemp=pd.json_normalize(dfw[i])
        if(is_position(dftemp, 0, poslist)):
            df=pd.concat([df, dftemp], ignore_index=True)       
    return df
```

```python
def process_stats(filename):
    df=pd.read_csv(filename)
    # Drop unnecessairy columns
    columns_to_drop = [col for col in df.columns if col.startswith('average.') and col != 'total.minutesOnField']

    df = df.drop(columns=columns_to_drop)
    columns_to_drop = [col for col in df.columns if col.startswith('percent.')]

    df = df.drop(columns=columns_to_drop)
    #exlude all players that played under 35 minutes
    df=df[df['total.minutesOnField']>=35]

    calc_col = [col for col in df.columns if col.startswith('total.') and col != 'total.minutesOnField']

    df[calc_col]=df[calc_col].div(df['total.minutesOnField']/90, axis=0)
    adjustedfilename='adjusted_'+filename
    #stats per 90
    df.to_csv(adjustedfilename, index=False)
    newfilename='normalized_'+filename

    mean_res = {col: df[col].mean() for col in calc_col}

    std_res = {col: df[col].std() for col in calc_col}
    #z-score of every category
    df[calc_col] = df[calc_col].apply(lambda col: (col - mean_res[col.name]) / std_res[col.name])
    df.to_csv(newfilename, index=False)
```

```python
#get all stats in form of a dataframe
path = 'data/stats/'
df=pd.read_csv('empty_df.csv')

statfilename='offmidstats.csv'
positionlist=offmid_pos

warnings.filterwarnings('ignore')

for filename in os.listdir(path):
    if filename.endswith('.json'):  # Check if the file is an json file
        file_path = os.path.join(path, filename)
        dftemp = readfile(file_path, positionlist)
        df=pd.concat([df, dftemp], ignore_index=True)

df.to_csv(statfilename, index=False)
#normalize
process_stats(statfilename)
        
```
